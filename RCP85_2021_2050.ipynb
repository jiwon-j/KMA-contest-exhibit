```python
import pandas as pd

file = pd.read_excel("RCP85_2021_2050.xlsx")

years = np.array(file["TIME"])
precip = np.array(file["RN"]*10)
```

```python
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm

path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=path, size=15).get_name()
plt.rc('font', family=font_name)

plt.figure(figsize=(10, 5))
plt.plot(years, precip, marker='o')
plt.xlim(2021, 2050)

z = np.polyfit(years, precip, 1)
p = np.poly1d(z)
plt.plot(years, p(years), "r--", color="red")
plt.grid(True)

plt.title("2021-2050 RCP8.5  ", fontsize=15, fontweight='bold')
plt.xlabel("", fontsize=13, fontweight='bold')
plt.ylabel(" (mm)", fontsize=13, fontweight='bold')

plt.show()
```

```python
rcp85 = pd.read_excel("RCP85_2021_2050.xlsx")

rcp_years = np.array( rcp85["TIME"] )
rcp_precip = np.array( rcp85["RN"]*10 )



file = pd.read_excel("1981_2020_total_precip.xlsx")

historical_years = np.array( file["year"] )
historical_precip = np.array( file["precip(mm)"] )


all_years = np.append(historical_years, rcp_years)
# print(all_years)
# print( len(all_years) )

all_precip = np.append( historical_precip, rcp_precip )
# print(all_precip)
# print( len(all_precip) )


plt.figure(figsize=(10, 5))
plt.plot(all_years, all_precip, linewidth=2, marker='o', markersize=4)


# plot trend
z = np.polyfit(all_years, all_precip, 1)
p = np.poly1d(z)
plt.plot(all_years, p(all_years), "r--", color="red", label="")
plt.legend(fontsize=12)


plt.grid(True)
plt.xlim(1981, 2050)
plt.ylim(800, 2000)
plt.title("1980-2020   + 2021-2050 RCP8.5  ", fontsize=15, fontweight='bold')
plt.xlabel("", fontsize=13, fontweight='bold')
plt.ylabel(" (mm)", fontsize=13, fontweight='bold')

plt.show()
```
