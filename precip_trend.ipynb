```python
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

file = pd.read_excel("1981_2020_total_precip.xlsx")

years = np.array( file["year"] )
precip = np.array( file["precip(mm)"] )


path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=path, size=15).get_name()
plt.rc('font', family=font_name)
    

def precip_trend(years, precip):
    
    plt.figure(figsize=(10, 5))
    plt.plot(years, precip, marker="o")
    plt.grid(True)

    # plot trend
    z = np.polyfit(years, precip, 1)
    p = np.poly1d(z)
    plt.plot(years, p(years), "r--", color="red", label="추세선")
    plt.legend(fontsize=12)
    
    plt.title("1981-2020 강수량 추세", fontsize=15, fontweight="bold")
    plt.xlabel("년도", fontsize=13, fontweight="bold")
    plt.ylabel("강수량 (mm)", fontsize=13, fontweight="bold")

    plt.show()
```

```python
precip_trend(years, precip)
```
