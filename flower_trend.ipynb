```python
def flower_trend(station, start_year, end_year):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    
    years = np.array(range(start_year, end_year+1))
    
    plt.figure(figsize=(10, 5))
    plt.title("cherry blossom blooming in "+station, fontsize=15)
    # ^ plot title first because the variable 'station' will be converted into numpy array
    
    file = pd.read_excel("flower.xlsx")
    station = file[station]
    station_data = np.array(station)

    # convert missing value to 0
    nan = np.isnan(station_data)
    station_data[nan] = 0
    plt.plot(years, station_data)
    
#     print("years :", years)
#     print(station_data)
    
    # plot trend
    z = np.polyfit(years, station_data, 1)
    p = np.poly1d(z)
    plt.plot(years, p(years), color="red")
    
    print(type(station))
```

```python
file_name = "flower.xlsx"
start_year = 1980
end_year = 2020
station = "Cheongju"

flower_trend(station, start_year, end_year)
```
